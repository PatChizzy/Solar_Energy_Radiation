# Exploaratory Data Analysis

```python
plt.figure(figsize=(8, 6))
plt.style.use('fivethirtyeight')
radiation_vs_month = data.loc[:, ['Daily_radiation', 'MonthPE']].groupby('MonthPE').sum()
radiation_vs_month.plot()
plt.xticks(data['MonthPE'], rotation=30)
plt.title('Total Solar Radiation by Month')
plt.show()
```

```python
f, ax = plt.subplots(figsize=(20, 5))
sns.lineplot(data= data, x="Date", y="Daily_radiation")

rad_avg= data['Daily_radiation'].mean()
ax.axhline(rad_avg, color='green', linestyle='--', label=f"Mean - {rad_avg:.2f}")

ax.set_axisbelow(True)
ax.minorticks_on()
ax.grid(which='major', linestyle='-', linewidth=0.5, color='white',)
ax.grid(which='minor', linestyle=':', linewidth=0.5, color='white', alpha=0.7)

plt.title('Daily Solar Radiation')
plt.show()
ax.legend()
```

```python
# Distribution of weather features and solar irradiation

f, ax = plt.subplots(1, 4, figsize=(20,4))
sns.histplot(data['Daily_radiation'], stat="density", bins=20, ax=ax[0])
sns.histplot(data['Daily_Pressure'],  stat="density", bins=20, ax=ax[1])
sns.histplot(data['Daily_Humidity'],  stat="density", bins=20, ax=ax[2])
sns.histplot(data['Daily_Temp'],      stat="density", bins=20, ax=ax[3])

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15, 6))
sns.barplot(x="MonthPE", y='Daily_radiation', data= data, palette= "BuPu")
ax.set_title('Daily Mean Radiation')
plt.show()
```

```python
# feature correlation
plt.figure(figsize=(10, 8))
corr= data.corr(method = "pearson")

sns.heatmap(corr, mask=np.zeros_like(corr, dtype=np.bool), cmap="coolwarm", square=True, annot=True, fmt=".2f")
plt.show()
```

## Statistical check to ensure the timeseries data is stationary

```python
# Adfuller test on target feature
from statsmodels.tsa.stattools import adfuller

radiation = data['Daily_radiation'].values
result = adfuller(radiation)

print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value)) # if p-value <= 0.05 the data is stationary, else its non stationary
```
